# Jockey Racing Stats Builder

Using the data derived from the features extracted by the [Feature Analysis](https://github.com/LeeSanderson/RacingData/blob/main/Data/FeatureAnalysis.ipynb) notebook, create a database (CSV file) of Jockey stats.

All stat computation is delegated to `race_analytics.features.jockey_stats.extract_jockey_stats` — this notebook no longer reimplements any win/place percentage or aggregation logic.

```python
import os

import pandas as pd

from race_analytics.data_path import get_data_path
from race_analytics.features.loader import load_stats
from race_analytics.features.jockey_stats import extract_jockey_stats

_DATA_PATH = get_data_path()
```

```python
# Load the race-level features produced by the FeaturePipeline (build_features / FeatureAnalysis).
races = load_stats(_DATA_PATH, "Race_Features.csv")
races["Off"] = pd.to_datetime(races["Off"])
```

```python
# Delegate all per-jockey stat computation to the shared features module.
jockey_stats = extract_jockey_stats(races)
jockey_stats.head()
```

```python
jockey_stats.to_csv(os.path.join(_DATA_PATH, "Jockey_Stats.csv"), index=False)
```
